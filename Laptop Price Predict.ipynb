```python
df_train = pd.read_csv('laptops_train.csv')
df_test = pd.read_csv('laptops_test.csv')
```

```python
def cat(df,cols):
    for col in cols:
        if df[col].dtype == 'O':
            df[col] = df[col].astype('category')
    return df
```

```python
cat(df,['Manufacturer', 'Model Name', 'Category', 'Screen Size', 'Screen',
       'CPU', 'RAM', ' Storage', 'GPU', 'Operating System', 'Weight'])
df.info()
```
